## What percent of black participants in the study graduated highschool?

```python
# Filtering

# Over 20
df_over_20 = df[df["RIDAGEYR"] >= 20]

# Over 20 and black
df_over_20_black = df_over_20[df_over_20["RIDRETH1"] == 4]

# Over 20, black and graduated high school
df_over_20_black_high = df_over_20_black[df_over_20_black["DMDEDUC2"].isin([3, 4, 5])]
```

```python
len(df_over_20_black_high) / len(df_over_20_black) * 100
```

## Show the average income levels of study participants by race in a bar graph. Which group has the highest income? The lowest?


```python
df_race_mean = df.groupby("RIDRETH3")["INDFMPIR"].mean().reset_index()
```

```python
# Grouping by race and taking mean of all columns
df_race_mean = df.groupby("RIDRETH3")["INDFMPIR"].mean().reset_index()

# Creating new column with the average income by multiplying the ratio with the income at poverty line
df_race_mean["AVGINCOME"] = df_race_mean["INDFMPIR"] * 14800
```

```python
df_race_mean
```

```python
df_race_mean.head()
```

```python
df_race_mean.replace(index=
                    {1: 'Mexican American',
                     2: 'Other Hispanic',
                     3: 'Non-Hispanic White',
                     4: 'Non-Hispanic Black',
                     6: 'Non-Hispanic Asian',
                     7: 'Other Race - Including Multi-Racial'
                    }, inplace = True)
```

```python
df_race_mean.reset_index(inplace = True)
```
