```python
print("The average age of the study participants is " + str(round(df[df["RIDAGEYR"] != '.']["RIDAGEYR"].mean())))
```

```python
# Get df of all black adults over 20
black = df["RIDRETH1"] == 4
over_twenty = df["RIDAGEYR"] >= 20
black_over_twenty = df[black & over_twenty]

# Get only those adults who graduated highschool
over_three = black_over_twenty["DMDEDUC2"] >= 3
under_five = black_over_twenty["DMDEDUC2"] <= 5
not_missing = black_over_twenty[under_five]
only_highschool = black_over_twenty[over_three & under_five]
print(str(len(only_highschool)/len(not_missing)*100) + "% of black adults surveyed graduated highschool")


```

```python
mean_values = df.groupby('RIDRETH1')['INDFMPIR'].mean().reset_index()
mean_values["INDFMPIR"] = mean_values["INDFMPIR"] * 14580
mean_values = mean_values.replace(1, "Mexican").replace(2, "Other Hispanic").replace(3, "White").replace(4, "Black").replace(5, "Other")

mean_values.plot(x='RIDRETH1', y='INDFMPIR', kind='bar', rot=0)
# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph Example')

# Show the plot
plt.show()
mean_values.head()
```
