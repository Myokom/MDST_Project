## Dropping columns & Cleaning 

```python
# DIABETES DROPPING
# List of columns to keep
columns_to_keep_diabetes = ['Respondent_ID', 
                            'Doctor_Told_Diabetes']

# Drop columns not in the list
df_diabetes = df_diabetes[columns_to_keep_diabetes]
```

```python
# DIET DROPPING
# List of columns to keep
columns_to_keep_diet = ["Respondent_ID",
                        "Meals_Outside",
                        "FastFood_Meals", 
                        "Ready_Eat_Foods_30D",
                        "Frozen_Meals_30D"]

# Drop columns not in the list
df_diet = df_diet[columns_to_keep_diet]
```

```python
# DEMO DROPPING
# List of columns to keep
columns_to_keep_demo = ['Respondent_ID', 
                            'Gender', 
                            'Age_Years', 
                            'Race']

# Drop columns not in the list
df_demo = df_demo[columns_to_keep_demo]
```

```python
# ACTIVITY DROPPING
# List of columns to keep
columns_to_keep_activity = ["Respondent_ID",
                            'Vigorous work activity',
                            'Walk or bicycle',
                            'Vigorous recreational activities']
# Drop columns not in the list

df_activity = df_activity[columns_to_keep_activity]
```

```python
# ALCOHOL DROPPING
# List of columns to keep
columns_to_keep_alc = ["Respondent_ID",
                       'Past 12 Months Frequency of Drinking Alcoholic Beverages']

# Drop columns not in the list
df_alc = df_alc[columns_to_keep_alc]

```

## Merging the datasets

```python
# Merge df_diabetes and df_diet on 'Respondent_ID'
df_merged = pd.merge(df_diabetes, df_diet, on='Respondent_ID')

# Merge df_demo on 'Respondent_ID' with the already merged DataFrame
df_merged = pd.merge(df_merged, df_demo, on='Respondent_ID')

# Merge 'df_activity' with the already merged DataFrame on 'Respondent_ID'
df_merged = pd.merge(df_merged, df_activity, on='Respondent_ID')

# Merge 'df_body' with the already merged DataFrame on 'Respondent_ID'
df_merged = pd.merge(df_merged, df_body, on='Respondent_ID')

# Merge 'df_alc' with the already merged DataFrame on 'Respondent_ID'
df_merged = pd.merge(df_merged, df_alc, on='Respondent_ID')

# Dropping ID column. Not predictive
df_merged = df_merged.drop('Respondent_ID', axis=1)
```

```python
# Define a small threshold value
threshold = 1e-10  # for example, numbers smaller than this are considered as zero

# Apply a function that sets values below the threshold to zero
df_merged = df_merged.applymap(lambda x: 0 if np.abs(x) < threshold else x)

# Now round the DataFrame
df = df_merged.round(0)
```

## Getting true values

```python
# Diabetes Cleaning

# Drop rows with value 9
df = df[df["Doctor_Told_Diabetes"] != 9]

# Convert 3's to 1's
df["Doctor_Told_Diabetes"].replace(3, 1, inplace = True)

# Replace values with descriptions
diabetes_value_to_description = {
    1: "Yes",
    2: "No"
}

df["Doctor_Told_Diabetes"].replace(diabetes_value_to_description, inplace = True)
```

```python
# Diet Cleaning

# Cleaning the specified column in the DataFrame
# Fill NA with mean


df["Meals_Outside"].replace({9999: None, 7777: None, 5555: 21}, inplace = True)
df["Meals_Outside"].fillna((round(df["Meals_Outside"].mean())), inplace=True)



df["FastFood_Meals"].replace({9999: None, 7777: None, 5555: 21}, inplace = True)
df["FastFood_Meals"].fillna((round(df["FastFood_Meals"].mean())), inplace=True)


df["Ready_Eat_Foods_30D"].replace({9999: None, 7777: None, 6666: 90}, inplace = True)
df["Ready_Eat_Foods_30D"].fillna((round(df["Ready_Eat_Foods_30D"].mean())), inplace=True)


df["Frozen_Meals_30D"].replace({9999: None, 7777: None, 6666: 90}, inplace = True)
df["Frozen_Meals_30D"].fillna((round(df["Frozen_Meals_30D"].mean())), inplace=True)

```

```python
# Demo Cleaning
df["Gender"].replace({1: "Male", 2: "Female"}, inplace=True)


# Create a mapping from the provided codes to the race descriptions
race_mapping = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other Race - Including Multi-Racial"
}

# Replace the codes in the 'Race' column with the descriptions
df['Race'].replace(race_mapping, inplace=True)
```

```python
# Activity Cleaning
# Replace 1 with 'Yes' and 2 with 'No'
df["Vigorous work activity"].replace({1: "Yes", 2: "No"}, inplace=True)

# Remove rows where 'Vigorous work activity' contains 7 or 9
df = df[(df["Vigorous work activity"] != 7) & (df["Vigorous work activity"] != 9)]


# Replace 1 with 'Yes' and 2 with 'No' for 'Walk or bicycle'
df["Walk or bicycle"].replace({1: "Yes", 2: "No"}, inplace=True)

# Remove rows where 'Walk or bicycle' contains 7 or 9
df = df[(df["Walk or bicycle"] != 7) & (df["Walk or bicycle"] != 9)]


# Replace 1 with 'Yes' and 2 with 'No' for 'Vigorous recreational activities'
df["Vigorous recreational activities"].replace({1: "Yes", 2: "No"}, inplace=True)

# Remove rows where 'Vigorous recreational activities' contains 7 or 9
df = df[(df["Vigorous recreational activities"] != 7) & (df["Vigorous recreational activities"] != 9)]

```

```python
# Body Meassurement Cleaning

# # Fill NAs with mean
df["Body Mass Index (kg/m**2)"].fillna(round(df["Body Mass Index (kg/m**2)"].mean()), inplace=True)
df["Waist Circumference (cm)"].fillna(round(df["Waist Circumference (cm)"].mean()), inplace=True)
df["Hip Circumference (cm)"].fillna(round(df["Hip Circumference (cm)"].mean()), inplace=True)

```

```python
# Alcohol Cleaning

df = df[(df['Past 12 Months Frequency of Drinking Alcoholic Beverages'] != 77) & 
        (df['Past 12 Months Frequency of Drinking Alcoholic Beverages'] != 99)]


# Calculate the rounded mean of the 'Past 12 Months Frequency of Drinking Alcoholic Beverages' column

# Fill NA values with the rounded mean of the column
df['Past 12 Months Frequency of Drinking Alcoholic Beverages'].fillna(round(df['Past 12 Months Frequency of Drinking Alcoholic Beverages'].mean()), inplace=True)


# Create a mapping from the provided codes to the descriptions
frequency_mapping = {
    0: "Never in the last year",
    1: "Every day",
    2: "Nearly every day",
    3: "3 to 4 times a week",
    4: "2 times a week",
    5: "Once a week",
    6: "2 to 3 times a month",
    7: "Once a month",
    8: "7 to 11 times in the last year",
    9: "3 to 6 times in the last year",
    10: "1 to 2 times in the last year"
}

# Replace the codes in the column with the descriptions
df['Past 12 Months Frequency of Drinking Alcoholic Beverages'].replace(frequency_mapping, inplace=True)


```
